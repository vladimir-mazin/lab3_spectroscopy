```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

Load measurement data

```python
filename = 'lif.txt'
_, rate, _, _, _, theta = np.genfromtxt(filename, unpack=True, skip_header=3)
```

```python
plt.plot(theta, rate)
#plt.yscale('log')
plt.grid()
peaks, _ =find_peaks(rate, height=80, threshold=10, prominence=50)
plt.plot(theta[peaks], rate[peaks],"x")

#plt.yscale('log')
plt.grid()
theta[peaks]
```
